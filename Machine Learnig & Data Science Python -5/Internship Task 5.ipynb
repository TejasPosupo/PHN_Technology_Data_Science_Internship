# 1. Write a Python code to Create a figure with 2x2 subplots.

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(2, 2)

axes[0, 0].plot([1, 2, 3, 4], [1, 4, 9, 16])
axes[0, 0].set_title('plot 1')

axes[0, 1].plot([1, 2, 3, 4], [1, 2, 3, 4])
axes[0, 1].set_title('plot 2')

axes[1, 0].plot([1, 2, 3, 4], [4, 3, 2, 1])
axes[1, 0].set_title('plot 3')

axes[1, 1].plot([1, 2, 3, 4], [16, 9, 4, 1])
axes[1, 1].set_title('plot 4')

plt.tight_layout()
plt.show()
```

# 2. In the first subplot (top-left), plot a bar chart using the following data:Categories: ['Category A', 'Category B', 'Category C', 'Category D', 'Category E'] Values: [23, 56, 41, 62, 19]

```python
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [23, 56, 41, 62, 19]

fig, axes = plt.subplots(2, 2)

# Bar Chart on the frist
axes[0, 0].bar(categories, values)
axes[0, 0].set_title('Bar Chart')
axes[0, 0].set_xlabel('Categories')
axes[0, 0].set_ylabel('Values')
axes[0, 0].tick_params(axis='x', rotation=90)

axes[0, 1].plot([1, 2, 3, 4], [1, 2, 3, 4])
axes[0, 1].set_title('plot 2')

axes[1, 0].plot([1, 2, 3, 4], [4, 3, 2, 1])
axes[1, 0].set_title('plot 3')

axes[1, 1].plot([1, 2, 3, 4], [16, 9, 4, 1])
axes[1, 1].set_title('plot 4')

plt.tight_layout()
plt.show()
```

```python
data = [12, 17, 21, 18, 14, 13, 16, 9, 12, 15, 19, 11, 14, 16, 20, 18, 15, 13, 16, 11, 10]

fig, axes = plt.subplots(2, 2)

# Bar Chart on the frist
axes[0, 0].bar(categories, values)
axes[0, 0].set_title('Bar Chart')
axes[0, 0].set_xlabel('Categories')
axes[0, 0].set_ylabel('Values')
axes[0, 0].tick_params(axis='x', rotation=90)

# Histogram on the second
axes[0, 1].hist(data, bins=5)
axes[0, 1].set_title('Histogram')
axes[0, 1].set_xlabel('Data')
axes[0, 1].set_ylabel('Frequency')

axes[1, 0].plot([1, 2, 3, 4], [4, 3, 2, 1])
axes[1, 0].set_title('plot 3')

axes[1, 1].plot([1, 2, 3, 4], [16, 9, 4, 1])
axes[1, 1].set_title('Splot 4')

plt.tight_layout()
plt.show()
```

# 3. In the third subplot (bottom-left), plot a pie chart using the following data: Labels: ['Apple', 'Banana', 'Orange', 'Mango'] Sizes: [30, 25, 15, 30]

```python
labels = ['Apple', 'Banana', 'Orange', 'Mango']
sizes = [30, 25, 15, 30]

fig, axes = plt.subplots(2, 2)

# Bar Chart on the frist
axes[0, 0].bar(categories, values)
axes[0, 0].set_title('Bar Chart')
axes[0, 0].set_xlabel('Categories')
axes[0, 0].set_ylabel('Values')
axes[0, 0].tick_params(axis='x', rotation=90)

# Histogram on the second
axes[0, 1].hist(data, bins=5)
axes[0, 1].set_title('Histogram')
axes[0, 1].set_xlabel('Data')
axes[0, 1].set_ylabel('Frequency')

# Pie Chart on the third
axes[1, 0].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title('Pie Chart')

axes[1, 1].plot([1, 2, 3, 4], [16, 9, 4, 1])
axes[1, 1].set_title('plot 4')

plt.tight_layout()
plt.show()
```
