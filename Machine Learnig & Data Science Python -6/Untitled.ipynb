```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler

# Step 1: Dataset Generation

# Define the number of equations in the dataset
num_equations = 100

# Generate random slopes and intercepts for the equations
slopes = np.random.uniform(low=-10, high=10, size=num_equations)
intercepts = np.random.uniform(low=-50, high=50, size=num_equations)

# Generate random input variables (x) within a specified range
x_range = (-10, 10)
x_values = np.random.uniform(low=x_range[0], high=x_range[1], size=(num_equations, 100))

# Generate the corresponding output values (y) for each equation
dataset = []
for i in range(num_equations):
    equation = {
        'slope': slopes[i],
        'intercept': intercepts[i],
        'x': x_values[i],
        'y': slopes[i] * x_values[i] + intercepts[i]
    }
    dataset.append(equation)

# Step 2: Data Pre-processing

# Perform data normalization on input variables (x) and output values (y)
scaler = MinMaxScaler()
for equation in dataset:
    equation['x'] = scaler.fit_transform(equation['x'].reshape(-1, 1)).flatten()
    equation['y'] = scaler.fit_transform(equation['y'].reshape(-1, 1)).flatten()

# Split the dataset into training and testing sets
train_data, test_data = train_test_split(dataset, test_size=0.2, random_state=42)

# Extract the features (input variables) and targets (output values) from the dataset
X_train = np.array([equation['x'] for equation in train_data])
y_train = np.array([equation['y'] for equation in train_data])

# Initialize and train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Perform predictions on test data
X_test = np.array([equation['x'] for equation in test_data])
y_test = np.array([equation['y'] for equation in test_data])
y_pred = model.predict(X_test)

# Evaluate the model performance
mse = np.mean((y_test - y_pred) ** 2)
print(f"Mean Squared Error: {mse:.4f}")
```
